```python
plt.plot(x,y,'b*--')
plt.xlabel("X Axis")
plt.ylabel("y Axis")
plt.title("Graph")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g*--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*--')
```

```python
f=plt.figure()
axes=f.add_axes([1,1,1,1])
axes.plot(x,y,'b*--')
```

```python

f=plt.figure()
axes1=f.add_axes([0.1,0.1,0.8,0.8])
axes2=f.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b*--')
axes1.set_xlabel("X Axis")
axes1.set_ylabel("y Axis")
axes1.set_title("Graph")

axes2.plot(y,x,'b*--')
axes2.set_xlabel("X Axis")
axes2.set_ylabel("y Axis")
axes2.set_title("Graph")
```

```python
fig,axes=plt.subplots(figsize=(12,3))
axes.plot(x,y,'r')
axes.set_xlabel("X Axis")
axes.set_ylabel("y Axis")
axes.set_title("Graph")
```
